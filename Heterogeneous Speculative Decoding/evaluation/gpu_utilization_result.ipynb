```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# gpu_utilization_api_calling = 'gpu_utilization_api_calling.txt'
# gpu_utilization_vanilla_sd = 'gpu_utilization_vanilla_sd.txt'
# gpu_utilization_mcsd = 'gpu_utilization_mcsd.txt'
gpu_utilization_mcsd = 'gpu_utilization_mcsd.txt'
gpu_utilization_mcsd_131 = 'gpu_utilization_mcsd_131.txt'
```

```python
gpu_utilization_131 = []
with open(gpu_utilization_mcsd_131, 'r') as file:
    gpu_utilization = list(map(int, file.readline().strip('[]').split(',')))
gpu_utilization_131 = np.array(gpu_utilization)*0.01
```

```python
gpu_utilization = []
with open(gpu_utilization_mcsd, 'r') as file:
    gpu_utilization = list(map(int, file.readline().strip('[]').split(',')))
gpu_utilization = np.array(gpu_utilization)*0.01
```

```python
# print(np.array(gpu_utilization).shape)
print(f"Total time in minute (3,2,1): {len(gpu_utilization)/60}")
print(f"Total time in minute (1,3,1): {len(gpu_utilization_131)/60}")
```

```python
# Create time array (data is collected every 100ms)
time = np.arange(0, len(gpu_utilization))
time_131 = np.arange(0, len(gpu_utilization_131))

plt.figure(figsize=(10, 12))
plt.subplot(2, 1, 1)
plt.plot(time, gpu_utilization)
plt.xlabel('Time (s)')
plt.ylabel('GPU Utilization (%)')
plt.title('GPU Utilization vs. Time')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(time_131, gpu_utilization_131)
plt.xlabel('Time (s)')
plt.ylabel('GPU Utilization (%)')
plt.title('GPU Utilization (131) vs. Time')
plt.grid(True)

plt.subplot(2, 1, 1)
plt.xlim(0, max(time.max(), time_131.max()))

plt.subplot(2, 1, 2)
plt.xlim(0, max(time.max(), time_131.max()))

plt.tight_layout()
plt.show()

# Calculate the integral of utilization and time
integral = integrate.simpson(gpu_utilization)
print(f"Integral of GPU utilization and time: {integral}")
```
