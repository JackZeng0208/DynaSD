```python
import pandas as pd
import matplotlib.pyplot as plt

file_with_decision_model = '/home/yixiaozeng/DynaSD/results/dynasd_acc_rate_speed_vs_width.csv'
file_without_decision_model = '/home/yixiaozeng/DynaSD/results/dynasd_acc_rate_speed_vs_width_no_decision_model.csv'

df_with_model = pd.read_csv(file_with_decision_model)
df_without_model = pd.read_csv(file_without_decision_model)

width_with_model = eval(df_with_model.iloc[0, 2])
acc_rate_with_model = eval(df_with_model.iloc[0, 3])
gen_speed_with_model = eval(df_with_model.iloc[0, 4])

width_without_model = eval(df_without_model.iloc[0, 2])
acc_rate_without_model = eval(df_without_model.iloc[0, 3])
gen_speed_without_model = eval(df_without_model.iloc[0, 4])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

axs[0].plot(width_with_model, acc_rate_with_model, label='With Decision Model', marker='o', color='orange')
axs[0].plot(width_without_model, acc_rate_without_model, label='Without Decision Model', marker='x', color='blue',linestyle='--')
axs[0].set_xlabel('Width')
axs[0].set_ylabel('Acceptance Rate')
# axs[0].set_title('Width vs Acceptance Rate')
axs[0].legend(loc="lower right")
axs[0].grid(True)

axs[1].plot(width_with_model, gen_speed_with_model, label='With Decision Model', marker='o', color='orange')
axs[1].plot(width_without_model, gen_speed_without_model, label='Without Decision Model', marker='x', color='blue', linestyle='--')
axs[1].set_xlabel('Width')
axs[1].set_ylabel('Generation Speed')
# axs[1].set_title('Width vs Generation Speed')
axs[1].legend(loc="lower right")
axs[1].set_ylim(60,85)
axs[1].grid(True)

plt.tight_layout()
plt.savefig('/home/yixiaozeng/DynaSD/results/decision_model_comparison.pdf')
```
