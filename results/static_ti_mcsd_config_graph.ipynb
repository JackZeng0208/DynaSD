```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("mcsd_results.csv")
```

```python
plt.figure(figsize=(10, 8))
speed_pivot = data.pivot_table(index='mcsd_config', values='Speed')
sns.heatmap(speed_pivot, annot=True, cmap="YlGnBu", fmt=".2f", cbar_kws={'label': 'Speed'})
plt.ylabel("MCSD Configuration")
plt.tight_layout()
plt.savefig("speed_heatmap.pdf")
plt.show()
```
